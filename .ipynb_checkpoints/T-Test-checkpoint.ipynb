```python
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats.weightstats import ttest_ind

data = pd.read_csv('imputed_data.csv', encoding = 'ISO-8859-1', index_col=0, header=0)
data.head()
```

```python
# Creating bins for the income column
def bins(x):
    if x < 25001:
        return 1
    if 25001 < x < 30001:
        return 2
    if 30001 < x < 45001:
        return 3
    if 45001 < x < 55001:
        return 4
    else:
        return 5
    
data['income_ranges'] = data['income'].apply(bins)  

print(data.income_ranges.value_counts())
```

```python
# Splitting the two independent variables into seperate dataframes
a = data['Y']
b = data['income_ranges']

# Using scipy's t-test function to calculate the t-statistic and p-value
# Place dataframe b before a to use b as the sample identifiers and a as the data
stats.ttest_ind(b, a, equal_var = False)
```
